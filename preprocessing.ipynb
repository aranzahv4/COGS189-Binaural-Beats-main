## Step 1: Loading Data

### Reading in Raw Data from repo

```python
!pip install mne

import numpy as np
import mne
import os
import gdown
import zipfile
from mne.preprocessing import ICA
from mne.time_frequency import tfr_morlet
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore', category=UserWarning, module='matplotlib')
plt.rcParams["figure.dpi"] = 150

%matplotlib inline

# Defining folder where data is stored
data_dir ="data/music_eeg/sub-01/ses-01"

# Loading the EEG data (eeg_raw.npy should contain raw signal)
eeg = np.load(os.path.join(data_dir, "eeg_raw.npy"))   # typical shape: channels x samples

# Defining sampling frequency (fs)
fs = 250

# Creating channel names automatically (should be 8 total + GND and REF)
ch_names = [f"Channel {i+1}" for i in range(eeg.shape[0])]

# definiing channel types (all EEG channels)
ch_types = ["eeg"] * eeg.shape[0]

# Creating an MNE Info object that stores metadata about the channels, sampling frequency/ rate, and channel types
info = mne.create_info(ch_names=ch_names, fs=fs, ch_types=ch_types)

# Converting the raw EEG data into an MNE Raw object, which is a standard format for storing and processing EEG data in MNE-Python
raw = mne.io.RawArray(eeg, info)

raw.filter(1., 40.)
raw.notch_filter(60.)
raw.set_eeg_reference("average")

raw.plot(duration=10, n_channels=eeg.shape[0], scalings="auto")
```
